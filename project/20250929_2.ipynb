#### 1. 문제 정의

- 현재 따릉이가 이용되고 있는 현황 조사
- 미래에 개선할 점에 대해서 분석 주제 정의
- 파이썬과 판다스 소개

<details>
<summary>따릉이: 서울시 자전거 대여 서비스</summary>

- 56% '출퇴근 시간대' 집중, 20분 이내 57%
- 대중교통 이용 전후 '퍼스트-라스트 마일' 문제 해결
- 도심보다 원거리 이동 많음
- 도심지 자전거 도로 등 인프라 확충
</details>

<details>
<summary>분석 주제 정의</summary>

- 언제 많이 타는지(이용건수)와 한 번 대여하면 어느 정도의 시간을 대여하는지(이용시간)
- 대여소가 위치한 장소적 특징에 따라서 사용자의 이용패턴
- 시간 개념과 대여소가 위치한 장소적 특징을 모두 고려한 사용자의 이용패턴
</details>

<details>
<summary>수업 환경</summary>

- 판다스(Pandas): 데이터 분석에 많이 사용되는 라이브러리
- 맷플랏립(Matplotlib): 파이썬으로 기본적인 차트들을 쉽게 그릴 수 있도록 도와주는 시각화 라이브러리
- 시본(Seaborn): matplotlib 기반으로 만들어진 통계 데이터 시각화 라이브러리로 high-level interface 제공
- 폴리움(Folium): 지도 시각화를 해주는 다양한 함수를 지닌 라이브러리
</details>
