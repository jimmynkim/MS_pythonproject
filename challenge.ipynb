지민이의 닉네임 my_nic과 적의 닉네임 enemy_nic이 문자열로 주어지고,  
지민이의 마나량 mana, 적의 체력 hp가 정수로 주어질 때

지민이가 남은 마나로 적을 쓰러뜨릴 수 있다면 "my_nic kill enemy_nic"을  
쓰러뜨릴 수 없다면 enemy_nic kill my_nic"을 출력하도록 코드를 작성해보세요. 

<br>
- 제한사항 

- 지민이의 캐릭터는 마나가 10보다 적으면 공격할 수 없습니다. 

- 1 ≤ my_nic의 길이, enemy_nic의 길이 ≤ 10

```python
my_nic = input("지민이의 닉네임을 설정하시오 (1~10자 이내): ")

if len(my_nic) < 1 or len(my_nic) > 10:
    print("지민이의 닉네임을 다시 설정해주세요. (1~10자 이내)")
    raise ValueError("닉네임 길이가 올바르지 않습니다. (1~10자 이내)") # 여기서 셀 실행 중단

enemy_nic = input("적의 닉네임을 설정하시오 (1~10자 이내): ")

if len(enemy_nic) < 1 or len(enemy_nic) > 10:
    print("적의 닉네임을 다시 설정해주세요. (1~10자 이내)")
    raise ValueError("닉네임 길이가 올바르지 않습니다. (1~10자 이내)")

mana = int(input("지민이의 마나량을 설정하시오 (1~1,000 이내): "))

if mana < 1 or mana > 1000:
    print("지민이의 마나량을 다시 설정하시오 (1~1,000 이내)")
    raise ValueError("마나량이 올바르지 않습니다. (1~1,000 이내)")

hp = int(input("적의 체력(hp)을 설정하시오 (1~1,000 이내):  "))

if hp < 1 or hp> 1000:
    print("적의 체력(hp)을 다시 설정하시오 (1~1,000 이내)")
    raise ValueError("적의 체력(hp)이 올바르지 않습니다. (1~1,000 이내)")

if mana*2 >= hp:
    print(f"{my_nic} kill {enemy_nic}")

else:
    print(f"{enemy_nic} kill {my_nic}")
```

### 12-1번

과일 가게를 운영해 주세요.  
f = ['사과', '바나나', '오렌지', '포도', '수박']   
c = [3, 3, 3, 3, 3] 



```python
f = ['사과', '바나나', '오렌지', '포도', '수박']
c = [3, 3, 3, 3, 3]

print("과일 가게에 오신 것을 환영합니다!")
print(f'오늘 판매하는 과일: {f}\n')

while sum(c) > 0:       # 모든 과일이 팔릴 때까지 반복
    print(f'현재 재고: {c}')

    fruit = input("어떤 과일을 구매하시겠습니까? (종료하려면 '그만'을 입력하세요)")
    if fruit == "그만":
        print("구매가 완료되었습니다.")
        break

    if fruit in f:
        idx = f.index(fruit)    # 과일 인덱스 찾기

        if c[idx] > 0: # 재고 확인
            c[idx] = c[idx] - 1
            print(f'{fruit} 을(를) 구매하셨습니다! 남은 {fruit}의 재고: {c[idx]}\n')
        else:
            print(f'죄송합니다. {fruit}의 재고가 없습니다.\n')


else: 
    print("모든 과일이 다 팔렸습니다. 가게를 종료합니다.")

```

### 12-2번

지민이는 햄버거 가게에서 아르바이트를 합니다.  

아래서부터, 빵 – 야채 – 고기 - 빵으로 쌓인 햄버거만 포장할수 있으면 재료는 순서대로 하나씩 들어와 쌓입니다.  

재료의 순서가 [야채, 빵, 빵, 야채, 고기, 빵, 야채, 고기, 빵]이면,  
지민이는 여섯 번째 재료가 쌓였을 때, 세 번째 재료부터 여섯 번째 재료를 이용하여 햄버거를 포장하고,  
아홉 번째 재료가 쌓였을 때, 두 번째 재료와 일곱 번째 재료부터 아홉 번째 재료를 이용하여 햄버거를 포장합니다.  
즉, 2개의 햄버거를 포장하게 됩니다. 

 

ingredient의 원소는 1, 2, 3 중 하나의 값이며, 순서대로 빵, 야채, 고기를 의미합니다. 

 

입출력 예 

ingredient                          / result 

[2, 1, 1, 2, 3, 1, 2, 3, 1]         / 2 

[1, 3, 2, 1, 2, 1, 3, 1, 2]         / 0 

```python
ingredient = [2, 1, 1, 2, 3, 1, 2, 3, 1]

burger = 0
stack = []

for i in ingredient:
    stack.append(i)

    # 스택에 재료가 4개 이상 쌓였고, 햄버거 순서([1, 2, 3, 1])와 일치하는 지 확인
    if len(stack) >= 4 and stack[-4:] == [1, 2, 3, 1]:
        burger = burger + 1
        # 포장된 햄버거 재료 4개 제거
        for _ in range(4):      # _는 값에 상관없이 다음 문장 반복
            stack.pop()

print(f'포장된 햄버거 개수: {burger}개')
```

```python
# 다른 방법

name = ['Alice', 'Bob', 'Charlie']
score = [[90, 85, 92],[78, 80, 88], [95, 92, 89]]

# 각 학생의 평균 점수 계산 및 출력
for i in range(len(name)):
    student_name = name[i]
    student_scores = score[i]
    average_score = sum(student_scores) / len(student_scores)
    print(f"{student_name}의 평균 점수: {average_score:.1f}")

print()

# 전체 평균 점수 계산 및 출력
total_scores = []
for student_scores in score:
    total_scores.extend(student_scores)
overall_average = sum(total_scores) / len(total_scores)
print(f"전체 학생 평균 점수: {overall_average}")

print()

# 90점 이상 점수를 가진 학생 이름 출력
high_achievers = []
for i in range(len(name)):
    if any(s >= 90 for s in score[i]):
        high_achievers.append(name[i])

print(f"90점 이상 점수가 있는 학생: {', '.join(high_achievers)}")
```
