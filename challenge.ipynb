(절사: 소수점 이하 또는 특정 자리수 이하의 숫자를 무조건 잘라내어 버리는 것)

### 7번

입력된 학생의 점수 score에 따라 성적을 A, B, C, D, F로 구분하여 출력하는 코드를 작성해주세요. 

A : 90 ≤ score 

B : 80 ≤ score < 90 

C : 70 ≤ score < 80 

D : 60 ≤ score < 70 

F : score < 60 

```python
score = int(input("점수를 입력해주시오: "))

if score >= 90:
    print(f"{score}의 성적은 A입니다.")
elif score >= 80:
    print(f"{score}의 성적은 B입니다.")
elif score >= 70:
    print(f"{score}의 성적은 C입니다.")
elif score >= 60:
    print(f"{score}의 성적은 D입니다.")
else:
    print(f"{score}의 성적은 F입니다.")
```

### 8-1번

지민이는 온라인 게임을 하던 도중 갑작스럽게 적을 만났습니다.

지민이의 캐릭터는 마나를 소모하며 공격하는 캐릭터이고,  
한 번 공격 시 마나 10을 소모하며 데미지는 20입니다.

지민이의 닉네임 my_nic과 적의 닉네임 enemy_nic이 문자열로 주어지고,  
지민이의 마나량 mana, 적의 체력 hp가 정수로 주어질 때

지민이가 남은 마나로 적을 쓰러뜨릴 수 있다면 "my_nic kill enemy_nic"을  
쓰러뜨릴 수 없다면 enemy_nic kill my_nic"을 출력하도록 코드를 작성해보세요. 

<br>
- 제한사항 

- 지민이의 캐릭터는 마나가 10보다 적으면 공격할 수 없습니다. 

- 1 ≤ my_nic의 길이, enemy_nic의 길이 ≤ 10 

- 1 ≤ mana, hp ≤ 1,000 

```python
my_nic = input("지민이의 닉네임을 설정하시오 (1~10자 이내): ")

if len(my_nic) < 1 or len(my_nic) > 10:
    print("지민이의 닉네임을 다시 설정해주세요. (1~10자 이내)")
    raise ValueError("닉네임 길이가 올바르지 않습니다. (1~10자 이내)") # 여기서 셀 실행 중단

enemy_nic = input("적의 닉네임을 설정하시오 (1~10자 이내): ")

if len(enemy_nic) < 1 or len(enemy_nic) > 10:
    print("적의 닉네임을 다시 설정해주세요. (1~10자 이내)")
    raise ValueError("닉네임 길이가 올바르지 않습니다. (1~10자 이내)")

mana = int(input("지민이의 마나량을 설정하시오 (1~1,000 이내): "))

if mana < 1 or mana > 1000:
    print("지민이의 마나량을 다시 설정하시오 (1~1,000 이내)")
    raise ValueError("마나량이 올바르지 않습니다. (1~1,000 이내)")

hp = int(input("적의 체력(hp)을 설정하시오 (1~1,000 이내):  "))

if hp < 1 or hp> 1000:
    print("적의 체력(hp)을 다시 설정하시오 (1~1,000 이내)")
    raise ValueError("적의 체력(hp)이 올바르지 않습니다. (1~1,000 이내)")

if mana*2 >= hp:
    print(f"{my_nic} kill {enemy_nic}")

else:
    print(f"{enemy_nic} kill {my_nic}")
```

```python
# 다른 방법

my_nic = input("나의 닉네임을 설정하시오: ")
enemy_nic = input("적의 닉네임을 설정하시오: ")
mana = int(input("나의 마나량을 설정하시오: "))
hp = int(input("적의 체력(hp)을 설정하시오: "))

att = mana // 10        # 공격 횟수
damage = att*20         # 적에게 입히는 데미지
remain_mana = mana%10   # 남은 마나
remain_hp = hp - damage

print("게임 결과")
print()
print(f"나의 닉네임: {my_nic}")
print(f"적의 닉네임: {enemy_nic}")
print(f"나의 마나량: {mana}")
print(f"적의 체력(hp): {hp}")
print(f"사용한 마나: {mana - remain_mana}")
print(f"남은 마나: {remain_mana}")
print(f"총 공격 횟수: {att}")
print(f"총 적에게 입힌 데미지: {damage}")
print(f"적에게 남은 체력: {remain_hp if remain_hp > 0 else 0}")
print()

if damage >= hp:
    print(f"게임 결과: {my_nic} kill {enemy_nic}")
else:
    print(f"게임 결과: {enemy_nic} kill {my_nic}")
```

### 10번

*을 가로로 a개, 세로로 b개 사용하여 직사각형을 만들려고 합니다.  
정수 a, b가 매개변수로 주어질 때, a x b 직사각형을 출력하도록 코드를 작성하세요 

```python
a = int(input("가로의 갯수를 입력하시오: "))
b = int(input("세로의 갯수를 입력하시오: "))

print(f"가로의 갯수: {a}, 세로의 갯수: {b}")

for j in range(0, b):
    print()
    for i in range(0, a):
        print("*", end="")
```
